```python
import pandas as pd
import numpy as np
import hvplot.pandas
import holoviews as hv
import plotly
```

```python
#


# loads airline dataframes
aal = pd.read_csv('Data/Airlines/AAL_quarterly_reports.csv', index_col='fiscalDateEnding', parse_dates=True, infer_datetime_format=True)
# aal without an index column set. hacky solution, but we can easily refer to aal_tmp['fiscalDateEnding'] as the date on graphs
aal_tmp = pd.read_csv('Data/Airlines/AAL_quarterly_reports.csv', parse_dates=True, infer_datetime_format=True)
alk = pd.read_csv('Data/Airlines/ALK_quarterly_reports.csv', index_col='fiscalDateEnding', parse_dates=True, infer_datetime_format=True)
dal = pd.read_csv('Data/Airlines/DAL_quarterly_reports.csv', index_col='fiscalDateEnding', parse_dates=True, infer_datetime_format=True)
luv = pd.read_csv('Data/Airlines/LUV_quarterly_reports.csv', index_col='fiscalDateEnding', parse_dates=True, infer_datetime_format=True)
ual = pd.read_csv('Data/Airlines/UAL_quarterly_reports.csv', index_col='fiscalDateEnding', parse_dates=True, infer_datetime_format=True)

# loads tech dataframes
aapl = pd.read_csv('Data/Tech/AAPL_quarterly_reports.csv', index_col='fiscalDateEnding', parse_dates=True, infer_datetime_format=True)
googl = pd.read_csv('Data/Tech/GOOGL_quarterly_reports.csv', index_col='fiscalDateEnding', parse_dates=True, infer_datetime_format=True)
amzn = pd.read_csv('Data/Tech/AMZN_quarterly_reports.csv', index_col='fiscalDateEnding', parse_dates=True, infer_datetime_format=True)
msft = pd.read_csv('Data/Tech/MSFT_quarterly_reports.csv', index_col='fiscalDateEnding', parse_dates=True, infer_datetime_format=True)
fb = pd.read_csv('Data/Tech/FB_quarterly_reports.csv', index_col='fiscalDateEnding', parse_dates=True, infer_datetime_format=True)

#sp500
spy = pd.read_csv('Data/SPY.csv', index_col='time', parse_dates=True, infer_datetime_format=True)
```

```python
# loads industries barsets
aal_bs = pd.read_csv('Data/AAL.csv', index_col='time', parse_dates=True, infer_datetime_format=True)
alk_bs = pd.read_csv('Data/ALK.csv', index_col='time', parse_dates=True, infer_datetime_format=True)
dal_bs = pd.read_csv('Data/DAL.csv', index_col='time', parse_dates=True, infer_datetime_format=True)
luv_bs = pd.read_csv('Data/LUV.csv', index_col='time', parse_dates=True, infer_datetime_format=True)
ual_bs = pd.read_csv('Data/UAL.csv', index_col='time', parse_dates=True, infer_datetime_format=True)

# loads tech dataframes
aapl_bs = pd.read_csv('Data/AAPL.csv', index_col='time', parse_dates=True, infer_datetime_format=True)
googl_bs = pd.read_csv('Data/GOOGL.csv', index_col='time', parse_dates=True, infer_datetime_format=True)
amzn_bs = pd.read_csv('Data/AMZN.csv', index_col='time', parse_dates=True, infer_datetime_format=True)
msft_bs = pd.read_csv('Data/MSFT.csv', index_col='time', parse_dates=True, infer_datetime_format=True)
fb_bs = pd.read_csv('Data/FB.csv', index_col='time', parse_dates=True, infer_datetime_format=True)
```

```python
# probably a bad way of doing this but we wanna put them all in one dataframe to easily plot with hvplot
data = {'date': alk_tmp['fiscalDateEnding'].array, 
        'aal': aal['totalCurrentAssets'].array,
        'alk': alk['totalCurrentAssets'].array, 
        'dal': dal['totalCurrentAssets'].array,
        'luv': luv['totalCurrentAssets'].array,
        'ual': ual['totalCurrentAssets'].array
       }
airline_total_assets_dataframe = pd.DataFrame(data=data)
truncate_after = airline_total_assets_dataframe.index.values[14]
airline_total_assets_dataframe.sort_values(by='date', ascending=True, inplace=True)

airline_total_assets_dataframe = airline_total_assets_dataframe.truncate(before=airline_total_assets_dataframe.index.values[14])
airline_total_assets_dataframe.set_index('date', inplace=True)

fig = airline_total_assets_dataframe.hvplot.line(rot=90, height=500, width=1000, grid=True).opts(yformatter="%.0f", title='Airline Assets')
fig
```

```python
cash_data = {'date': alk_tmp['fiscalDateEnding'].array, 
        'aal': aal['cash'].array,
        'alk': alk['cash'].array, 
        'dal': dal['cash'].array,
        'luv': luv['cash'].array,
        'ual': ual['cash'].array
       }
airline_cash_df = pd.DataFrame(data=cash_data)
truncate_after = airline_cash_df.index.values[14]
airline_cash_df.sort_values(by='date', ascending=True, inplace=True)

airline_cash_df = airline_cash_df.truncate(before=airline_cash_df.index.values[14])
airline_cash_df.set_index('date', inplace=True)

fig = airline_cash_df.hvplot.line(rot=90, height=500, width=1000, grid=True).opts(yformatter="%.0f", title='Airline Cash')
fig
```

```python
sp500_close = spy['close'].pct_change().to_frame()

airline_barsets_pct_change = {
    'date': alk_bs.index.values,
    'alk': alk_bs['Close'].pct_change().array,
    'ual': ual_bs['Close'].pct_change().array,
    'luv': luv_bs['Close'].pct_change().array,
    'dal': dal_bs['Close'].pct_change().array,
    'aal': aal_bs['Close'].pct_change().array
}
airline_barsets_pct_change = pd.DataFrame(data=airline_barsets_pct_change)
airline_barsets_pct_change.set_index('date', inplace=True)
airline_barsets_pct_change.dropna(inplace=True)

tech_barsets_pct_change = {
    'date': aapl_bs.index.values,
    'aapl': aapl_bs['Close'].pct_change().array, 
    'googl': googl_bs['Close'].pct_change().array, 
    'amzn': amzn_bs['Close'].pct_change().array, 
    'msft': msft_bs['Close'].pct_change().array, 
    'fb': fb_bs['Close'].pct_change().array
}
tech_barsets_pct_change = pd.DataFrame.from_dict(tech_barsets_pct_change)
tech_barsets_pct_change.set_index('date', inplace=True)
tech_barsets_pct_change['Tech Percent Change'] = tech_barsets_pct_change[['aapl', 'googl', 'amzn', 'msft', 'fb']].mean(axis=1)

airline_barsets_pct_change['Airline Percent Change'] = airline_barsets_pct_change[['alk', 'ual', 'luv', 'dal', 'aal']].mean(axis=1)



sp500_close = sp500_close.truncate(before=sp500_close.index.values[542])
airline_barsets_pct_change = airline_barsets_pct_change.truncate(before=airline_barsets_pct_change.index.values[542])
tech_barsets_pct_change = tech_barsets_pct_change.truncate(before=tech_barsets_pct_change.index.values[542])
#airline_barsets_pct_change.truncate(before=sp500_close.index.values[542])
#tech_barsets_pct_change.truncate(before=sp500_close.index.values[542])


sp500_close.columns = ['S&P500 Percent Change']
fig1 = sp500_close['S&P500 Percent Change'].hvplot(rot=90, height=500, width=1000, grid=True).opts(yformatter="%.0f", title='Tech & Airline Percentage Change vs S&P500')
fig2 = airline_barsets_pct_change['Airline Percent Change'].hvplot()
fig3 = tech_barsets_pct_change['Tech Percent Change'].hvplot()
#fig = fig1 * fig2 * fig3
fig1 * fig2 * fig3
```
